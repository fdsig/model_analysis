```python
import json 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from google_drive_downloader import GoogleDriveDownloader as gdd

```

```python
result_fids = [i for i in os.scandir('vit_fb_logs/results/')]
results_dict =  { }
for fid_ in result_fids:
    with open(fid_, 'r') as fid:
        text = fid.readlines()
        key = fid_.name.split('.')[0]
        print(key)
        results_dict[key]=text
```

```python
 results = {
     i:{'epoch_'+str(json.loads(j)['epoch']):json.loads(j) for j in results_dict[i]} 
     for i in results_dict
 }
results.keys()
```

```python
deep_keys =  [list(results[i][j].keys() )for i in results for j in list(results[i].keys())][0]
deep_keys[:4]
```

```python
metrics = {deep_keys[k]:{i:np.array([results[i][j][deep_keys[k]] 
           for j in results[i]][:5]) 
        for i in results} for k in range(1,4)}

train_loss, test_loss, test_acc = (metrics[i] for i in metrics)
test_acc = {key:test_acc[key]/100 for key in test_acc}
metrics['test_acc1']=test_acc
maxs = {j:max([np.round(metrics[j][i][metrics[j][i].argmax()],2) for i in metrics[j]]) for j in metrics
}
mins = {j:min([np.round(metrics[j][i][metrics[j][i].argmin()],2) for i in metrics[j]]) for j in metrics
}
mins
```

```python
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['figure.dpi'] = 200
plt.tight_layout()
dim = 3
fig, ax = plt.subplots(3,1,figsize=(20,30))
fig.patch.set_facecolor('xkcd:white')
coords = [[i,0] for i in range(dim)]
metric_types = [[i,j]for i in ['train_loss','validation'] for j in ['acc', 'loss']]

legend = list(metrics.keys())
for idx,key in enumerate(metrics):
    metric = metrics[key]
    legend = list(metric.keys())
    print(metric)
    coord = coords[idx]
    low_, high_ = 0,0
    for i in metric:
        low, high = (metric[i][np.argmin(metric[i])], 
                     metric[i][np.argmax(metric[i])])
        print(low)
        if low>low_:
            low_=low
        if high>high_:
            high_= high
        
        ax[coord[0]].plot(range(1,len(metric[i])+1),metric[i])
        ax[coord[0]].grid(which='both')
        ax[coord[0]].grid(which='minor',alpha=0.3)
        ax[coord[0]].grid(which='major',alpha=0.9)
        ax[coord[0]].set_yticks(np.linspace(mins[key],maxs[key],10))
        ax[coord[0]].set_xticks(np.linspace(1,5,5))
        ax[coord[0]].set_ylabel(key, fontsize=25)
        ax[coord[0]].set_xlabel('Epoch', fontsize=25)
        ax[coord[0]].legend(legend,loc=(4), fontsize=20)
        ax[coord[0]].set_title(key, fontsize=30)
    fig.suptitle('Model Metrics Compared', fontsize=40)
```
